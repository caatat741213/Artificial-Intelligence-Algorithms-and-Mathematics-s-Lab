4.Create a 2 D array [[10,20,45], [30,12,16], [42,17,56]] and perform the following operations:

Slice the 2D array to get the first two rows, slice the 2D array to get the last two rows.


```python
arr_2d = np.array([[10, 20, 45], [30, 12, 16], [42, 17, 56]])
first_two_rows = arr_2d[:2, :]
last_two_rows = arr_2d[1:, :]

print("original array:\n", arr_2d)
print("first_two_rows:\n", first_two_rows)
print("last two rows:\n", last_two_rows)
```

5.Create two 2x2 arrays and demonstrate how you can stack the elements vertically, horizontally, and split the arrays into smaller arrays.

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

# Vertical Stack
v_stack = np.vstack((a, b))

# Horizontal Stack
h_stack = np.hstack((a, b))

# Split
split_arr = np.hsplit(h_stack, 2)

print("Vertical Stack:\n", v_stack)
print("Horizontal Stack:\n", h_stack)
print("Split Array:\n", split_arr)
```

6.Create two matrices X= ([[5, 7, 2], [4, 5, 6], [7, 4 ,2]]) Y= ([[4, 2], [6, 2], [4, 2]]), Is it possible to multiply these matrices?

Demonstrate the case when it is not possible to.


```python
X = np.array([[5, 7, 2], [4, 5, 6], [7, 4, 2]]) # 3x3
Y = np.array([[4, 2], [6, 2], [4, 2]])       # 3x2

print("Matrix multiplication rule: The number of rows in the left matrix must equal the number of columns in the right matrix.")
product = np.dot(X, Y)
print("X*Y:\n", product)
print("So Y(3x2)*X(3x3)==>Error")
```

7.Create two arrays, x = ([2, -1, -8]) y = ([3, 1, -2]), Find the Shape, Number of dimensions of vector x.
         
Reshape the vector x to a matrix of size (3,1) and determine the number of dimensions after reshaping y to a matrix of (3,1).


```python
x = np.array([2, -1, -8])
y = np.array([3, 1, -2])

print(f"x shape: {x.shape}, dimensions: {x.ndim}")

# Reshape
x_reshaped = x.reshape(3, 1)
y_reshaped = y.reshape(3, 1)
print(x_reshaped)
print(y_reshaped)

print(f"new x shape: {x_reshaped.shape}, dimensions: {x_reshaped.ndim}")
```
